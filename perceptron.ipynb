```python
import random
```

```python
data=[  
    [2, 8, 1],  # Study Hours = 2, Sleep Hours = 8, Pass = 1
    [1, 6, 0],  
    [3, 7, 1],  
    [0, 5, 0]   
]

w1=random.random()
w2=random.random()
b=random.random()
lr=0.1
```

```python
def activation(x):
    return 1 if x>=0 else 0


def perceptron(data,w1,w2,b,lr,epoch=100):
    for epoch in range(epoch):
     for row in data:
        x1=row[0]
        x2=row[1]
        y=row[2]

        weighted=x1*w1+x2*w2+b
         
        prediction=activation(weighted)
        error=y-prediction

        w1 +=lr*error*x1
        w2 +=lr*error*x2
        b +=lr*error

        print(f'Epoch {epoch+1}: Weights: w1={w1:.4f}, w2={w2:.4f}, b={b:.4f},weighted={b:.4f}')

    return w1, w2, b

```

```python
w1, w2, b = perceptron(data, w1, w2, b, lr)
```

```python
def test(x1, x2, w1, w2, b):
    pred = x1 * w1 + x2 * w2 + b
    return 1 if pred >= 0 else 0


x1, x2 = 1, 6  
prediction = test(x1, x2, w1, w2, b)
print(f"Prediction: {prediction}")
print(x1, x2, w1, w2, b)
```
