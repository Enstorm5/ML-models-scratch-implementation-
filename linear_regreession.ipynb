```python

x = [1,2,3,4,5]
y = [2,3,4,5,6]

totalxy = 0
for i in range(len(x)):
    totalxy += (x[i] * y[i])
print(totalxy)

squared=0
for i in range(len(x)):
  squared+=(x[i]*x[i])
print(squared)

x_sum = sum(x)
y_sum = sum(y)
print(x_sum)
print(y_sum)
```

```python
slope=(len(x)*totalxy - x_sum*y_sum)/((len(x)*squared)-(x_sum*x_sum))
print(slope)
```

```python
intercept=(y_sum/len(y))-slope*(x_sum/len(x))

print(intercept)
```
